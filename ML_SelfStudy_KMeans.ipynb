Among Unsupervised Machine Learning Algorithms, **K means clustering** is one of the most
popular. It is used to solve Classification Problems. K Means clustering divides the unlabelled
data into different groups, called clusters, based on features and patterns that are similar to each
other.

Implement **K Means Clustering** Algorithm for a randomly generated sequence.

* Import the basic libraries


```python
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import random
import numpy as np
```

* Generate 200 random numbers in 2 dimensional space


```python
randset=[]
i=0
while(i<200):
    x=random.randint(-50,50)
    y=random.randint(-50,50)
    if (x,y) in randset: continue
    randset.append((x,y))
    i+=1
X=[]
Y=[]
for i in range(len(randset)):
    X.append(randset[i][0])
    Y.append(randset[i][1])
x=np.array(randset)
```

* Train K Means algorithm by considering number of clusters equal to 3


```python

kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
```

* Display the 3 clusters on 2D scattered plot.



```python
plt.scatter(x[:,0],x[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0] ,kmeans.cluster_centers_[:,1], color='black')
plt.title("Scattered Plot for 3 Clusters")
plt.show()
```

* Use Elbow Curve method to choose the number of centroids for improved clustering from the
poor data we generated in step 1 by considering computational expense and knowledge gain
from a dataset.


```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method using Inertia')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

* Mention the optimal number of clusters chosen and display improved clusters on 2D
scattered plot

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(x)
plt.scatter(x[:,0],x[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0] ,kmeans.cluster_centers_[:,1], color='black')
plt.title("Scattered Plot for 4 Clusters(Optimal Number)")
plt.show()
```
